#  HW 3: Assignment 1 - Distributions

```python
from __future__ import print_function
__author__= 'Manushi'
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit, minimize
from scipy.stats import norm

%matplotlib inline
```

```python
distributions = {} #Normal, Poisson, Binomial, Chi - squared, Gamma, Laplace distributions are chosen
```

```python
populationMean = 10
sd = 1
```

```python
#Sample space Size, 10 < N < 2000

sampleSpaceSize = 2000/(np.array(range(1,100)+[10]))
print (sampleSpaceSize, sampleSpaceSize.shape)
```

```python
distributions['normal'] = {}
for n in sampleSpaceSize:
    distributions['normal'][n] = np.random.normal(populationMean, sd, size=n)
means={}
means['normal'] = {}
figr = plt.figure(figsize = (10,10))
axNormal = figr.add_subplot(1,1,1)
for nn in distributions['normal']:
    means['normal'][nn] = distributions['normal'][nn].mean()
    axNormal.plot(nn, means['normal'][nn], 'o')
    axNormal.set_xlabel('Sample Size')
    axNormal.set_ylabel('Sample Mean')
    axNormal.plot([min(sampleSpaceSize), max(sampleSpaceSize)],[populationMean,populationMean], 'k')
    axNormal.grid()
```

```python
allMeans = []
for n in means['normal']:
    allMeans.append(means['normal'][n])
```

## 2. Poisson Distribution

```python
distributions['poisson'] = {}
for n in sampleSpaceSize:
    distributions['poisson'][n] = np.random.poisson(populationMean, size=n)
means['poisson'] = {}
figr = plt.figure(figsize = (10,10))
axPoisson = figr.add_subplot(1,1,1)
for nn in distributions['poisson']:
    means['poisson'][nn] = distributions['poisson'][nn].mean()
    axPoisson.plot(nn, means['poisson'][nn], 'o')
    axPoisson.set_xlabel('Sample Size')
    axPoisson.set_ylabel('Sample Mean')
    axPoisson.plot([min(sampleSpaceSize), max(sampleSpaceSize)],[populationMean,populationMean], 'k')
    axPoisson.grid()
```

```python
for n in means['poisson']:
    allMeans.append(means['poisson'][n])
```

## 3. Binomial Distribution

```python
distributions['binomial'] = {}
for n in sampleSpaceSize:
    distributions['binomial'][n] = np.random.binomial(20, p=0.5, size=n)
means['binomial'] = {}
figr = plt.figure(figsize = (10,10))
axBinomial = figr.add_subplot(1,1,1)
for nn in distributions['binomial']:
    means['binomial'][nn] = distributions['binomial'][nn].mean()
    axBinomial.plot(nn, means['binomial'][nn], 'o')
    axBinomial.set_xlabel('Sample Size')
    axBinomial.set_ylabel('Sample Mean')
    axBinomial.plot([min(sampleSpaceSize), max(sampleSpaceSize)],[populationMean,populationMean], 'k')
    axBinomial.grid()
```

```python
for n in means['binomial']:
    allMeans.append(means['binomial'][n])
```

## 4. Chi - squared distribution

```python
distributions['chi-squared'] = {}
for n in sampleSpaceSize:
    distributions['chi-squared'][n] = np.random.chisquare(populationMean, size=n)
means['chi-squared'] = {}
figr = plt.figure(figsize = (10,10))
axChisquare = figr.add_subplot(1,1,1)
for nn in distributions['chi-squared']:
    means['chi-squared'][nn] = distributions['chi-squared'][nn].mean()
    axChisquare.plot(nn, means['chi-squared'][nn], 'o')
    axChisquare.set_xlabel('Sample Size')
    axChisquare.set_ylabel('Sample Mean')
    axChisquare.plot([min(sampleSpaceSize), max(sampleSpaceSize)],[populationMean,populationMean], 'k')
    axChisquare.grid()
```

```python
for n in means['chi-squared']:
    allMeans.append(means['chi-squared'][n])
```

## 5. Gamma Distribution

```python
distributions['gamma'] = {}
for n in sampleSpaceSize:
    distributions['gamma'][n] = np.random.gamma(shape=5, scale =5, size=n)
means['gamma'] = {}
figr = plt.figure(figsize = (10,10))
axGamma = figr.add_subplot(1,1,1)
for nn in distributions['gamma']:
    means['gamma'][nn] = distributions['gamma'][nn].mean()
    axGamma.plot(nn, means['gamma'][nn], 'o')
    axGamma.set_xlabel('Sample Size')
    axGamma.set_ylabel('Sample Mean')
    axGamma.plot([min(sampleSpaceSize), max(sampleSpaceSize)],[15,15], 'k')
    axGamma.grid()
```

```python
for n in means['gamma']:
    allMeans.append(means['gamma'][n])
```

## 6. Laplace Distribution

```python
distributions['laplace'] = {}
for n in sampleSpaceSize:
    distributions['laplace'][n] = np.random.laplace(loc=populationMean, scale =5, size=n)
means['laplace'] = {}
figr = plt.figure(figsize = (10,10))
axLaplace = figr.add_subplot(1,1,1)
for nn in distributions['laplace']:
    means['laplace'][nn] = distributions['laplace'][nn].mean()
    axLaplace.plot(nn, means['laplace'][nn], 'o')
    axLaplace.set_xlabel('Sample Size')
    axLaplace.set_ylabel('Sample Mean')
    axLaplace.plot([min(sampleSpaceSize), max(sampleSpaceSize)],[populationMean,populationMean], 'k')
    axLaplace.grid()
```

```python
for n in means['laplace']:
    allMeans.append(means['laplace'][n])
```

### DISTRIBUTION OF SAMPLE MEANS

```python
plt.figure(figsize = (8,8))
plt.hist(allMeans,bins=20)
plt.title('Normal distribution of sample means')
plt.xlabel('Sample Mean')
plt.ylabel('N')
plt.grid()
plt.show
```
